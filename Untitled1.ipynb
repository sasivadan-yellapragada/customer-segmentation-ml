```python
import pandas as pd
df = pd.read_csv("bestsellers with categories_final.csv")  # Use the exact filename
df.head()  # Display the first few rows
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns



# Select relevant features
features = ['User Rating', 'Reviews', 'Price', 'Genre']
data = df[features].copy()

# Encode categorical variable (Genre)
le = LabelEncoder()
data['Genre'] = le.fit_transform(data['Genre'])

# Standardize numerical features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['User Rating', 'Reviews', 'Price']])

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(data_scaled)

# Visualize the clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['User Rating'], y=data['Reviews'], hue=data['Cluster'], palette='viridis')
plt.title("Customer Segmentation based on Book Features")
plt.xlabel("User Rating")
plt.ylabel("Reviews")
plt.show()

# Add clusters back to the original dataset
df['Cluster'] = data['Cluster']

# Save the segmented data
df.to_csv("bestsellers_clustered.csv", index=False)

print("Customer segmentation completed and saved!")
```

```python
# Select only numeric columns for grouping
numeric_cols = df.select_dtypes(include=['number'])  # Get numeric columns only

cluster_summary = numeric_cols.groupby(df['Cluster']).mean()

print(cluster_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='Cluster', palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Cluster', y='Price', data=df, palette='coolwarm')
plt.title('Price Distribution Across Clusters')
plt.show()
```
